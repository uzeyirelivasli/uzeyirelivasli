```python
# Input: nums = [-1,0,3,5,9,12], target = 9
# Output: 4
# Explanation: 9 exists in nums and its index is 4

nums = [-1,0,3,5,9,12]
target = 9

class Solution:
    def search(self, nums, target: int):
        
        left = 0
        right = len(nums) - 1

        

        while left <= right:
       
            middle = (left + right) // 2
            
            if nums[middle] == target:
                return middle

            if nums[middle] < target:
                left = middle + 1
            else:
                right = middle - 1
                
        return -1 
            

        
print(Solution.search('',nums,target))
        
        
```

```python
# filter(function,list)

# 1. Önce veri setimizi (sözlüklerden oluşan liste) tanımlayalım
urunler = [
    {"ad": "Kulaklık", "fiyat": 250},
    {"ad": "Mouse", "fiyat": 80},
    {"ad": "Klavye", "fiyat": 120},
    {"ad": "Mousepad", "fiyat": 40}
]

ucuz_urunler = list(filter(lambda urun: urun["fiyat"] < 100, urunler))

print(ucuz_urunler)
# Çıktı: [{'ad': 'Mouse', 'fiyat': 80}, {'ad': 'Mousepad', 'fiyat': 40}]
```

1. == (Equality)
What does it check? Whether the values (content) are the same.
The Question: "Are the things inside these two boxes the same?"
2. is (Identity)
What does it check? Whether the objects are the same location (address) in memory.
The Question: "Are these two variables actually the exact same object?"
A Quick Example to See the Difference:
python
list_a = [1, 2, 3]
list_b = [1, 2, 3]
list_c = list_a

# VALUE CHECK (==)
print(list_a == list_b) # True (Because the numbers inside are the same)

# IDENTITY CHECK (is)
print(list_a is list_b) # False (Because they are two different lists in memory, they just look alike)
print(list_a is list_c) # True (Because 'list_c = list_a' makes them point to the exact same spot)
