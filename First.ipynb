```python
# define func that takes list as an argument. iterate every item and
# if item < 20 add to total
def count(numbers_list):
    total = 0
    for number in numbers_list:
        if number < 20:
            total += number
    return total
exercise_list = [10,40,30,15]
count(exercise_list)
```

```python
# Iterating over Dictionaries
prices = {
    'box_of_spagetti' : 4,
    'lasagna' : 5,
    'hamburger' : 2
}
quantity = {
    'box_of_spagetti' : 6,
    'lasagna' : 10,
    'hamburger' : 0
}

money_spent = 0
five_bucks_piece = 0

for i in prices :
    money_spent += prices[i] * quantity[i]
    if prices[i] >= 5:
        five_bucks_piece += prices[i] * quantity[i]
print(money_spent)
print(five_bucks_piece)
```
